Handling Duplicates

```python
'''
In our case, we don't have duplicate rows, so we will apppend the dataset to itself
'''
csv_df_temp = csv_df.append(csv_df)
csv_df_temp.shape
```

```python
# Remove the duplicates using drop_duplicates method

csv_df_temp = csv_df_temp.drop_duplicates()
csv_df_temp.shape
```

```python
''' In the above method, we are required to put an "=" sign to drop the duplicates.
    We can use "inplace = true" inside the above method to make the function drop duplicates in itself without using "=" sign
'''
csv_df_temp = csv_df.append(csv_df)
print(csv_df_temp.shape)
csv_df_temp.drop_duplicates(inplace = True)
csv_df_temp.shape
```

```python
''' Apart from "inplace", another important argument for drop_duplicates is keep and it has three values :-

    df.drop_duplicates(keep = first) will drop all the duplicates except the first occurence of that value
    df.drop_duplicates(keep = last) will drop all the duplicates except the last occurence of that value
    df.drop_duplicates(keep = False) will drop all the occurences of that value
'''
csv_df_temp = csv_df.append(csv_df)
print(csv_df_temp.shape)
csv_df_temp.drop_duplicates(inplace = True, keep = False)
csv_df_temp.shape
```

```python
''' Imputation
When we have more null values in a dataset and since, dropna() method drops the complete row/columnn at once, we
follow the technique of filling these null values with their column's/row's mean/variance, etc instead of dropping them.
This technique is called imputation'''

#Store a column in a list variable
patient_nbr = csv_df["patient_nbr"]
patient_nbr.head()
#patient_nbr
```

```python
#Calculate column's mean

patient_nbr_mean = patient_nbr.mean()
patient_nbr_mean
```

```python
#Now fill back the null values in this column with the mean value

csv_df.fillna(patient_nbr_mean, inplace = True)
```

```python
# we can check again to see that the null values have been filled. 
csv_df.isnull().sum()
```

**Applying Functions for different Operations over Dataframe**

```python
''' Making individual changes for large dataframes is a hideous task and hence we can write functions for these operations 
and can apply these functions over the dataframe using df.apply() function, as follows'''

# Create a function that return "fit" if the patient's insulin is steady, else return "unfit"

def health_function(insulin):
  if insulin == "Steady":
    return "fit"
  else:
    return "unfit"
```

```python
''' Now pass the above function as an argument in the apply function and store the results 
in a new column of dataframe named "health_status" '''

csv_df["health_status"] = csv_df["insulin"].apply(health_function)
csv_df["health_status"]
```

```python
####  We can perform the similar task using a LAMBDA function as well 

csv_df["health_status"] = csv_df["insulin"].apply(lambda insulin : "fit" if insulin == "Steady" else "unfit")
csv_df["health_status"]
```
